## Clustering

In order to get ratings data to look at, I will need to run it through a custom clustering model which clusters by geographical location using latitude, longitude and google summary_polyline format. Let's see how this works. First I will create a dataframe on which to predict clusters.

```python
df = large_df[large_df['commute'] == 0] # drop all where commute = true and store as new dataframe, df
df = df[df['state'] == 'Colorado'] # drop all where state != Colorado
df = df[df['type'] == 'Ride'] # drop everything that is not of type 'Ride' for now
df.dropna(subset=['map_summary_polyline'], inplace=True) # drop all where map_summary_polyline is nan
df.reset_index(drop=True, inplace=True)
```

Next build training and test set. Since this is clustering, it is unsupervised, therefor there are no labels to test against. The test set is simply to test the transform method on held out or new data. I'm testing a set of 5 random athletes to make sure everything works as expected.

```python
sample_athlete_ids = df.athlete_id.sample(5, random_state=42) # choose 5 random athlete_ids

mask = df.athlete_id.isin(sample_athlete_ids)
train = df[~mask].reset_index(drop=True)
test = df[mask].reset_index(drop=True)
```

```python
test.shape
```

Next start the clusterer, fit it on training data, predict labels for the training set, then transform the test set:

```python
import sys
sys.path.insert(0, '../velosparkapp/recommender/')
from clustering import PolylineClusterMaker

clusterer = PolylineClusterMaker()
clusterer.fit(train)
train_labels = clusterer.predict(train)
test_labels, centroids = clusterer.transform(test)
```

## ALS

Apache Spark ML implements alternating least squares (ALS) for collaborative filtering, a very popular algorithm for making recommendations.

ALS recommender is a matrix factorization algorithm that uses Alternating Least Squares with Weighted-Lamda-Regularization (ALS-WR). It makes the regularization parameter less dependent on the scale of the dataset. This way the best parameter learned from the sampled subset can be aplied to the full dataset and we will get similar performance. The latent factors should explain the observed athlete to activity ratings and map new athletes to optimal activity recommendations.

## Considerations

The ALS model used assumes, by default, that the ratings are explicit. In this case the ratings are inferred from other signals. The parameter implicitPrefs should be set to True to yield better results.

Implicit feedback data has a number of distinctions to consider, which prevent the use of algorithms designed for explicit feedback. Prime characteristics are as follows:

1. No negative feedback. By observing activity behavior for different users, we can infer which activities they probably like and thus chose to spend time doing. It is hard to reliably infer which activities a user did not like versus those they weren't able to do for other reasons. This assymetric property of the data doesn't occur with explicit feedback where the user states what they prefer and what they do not prefer.

2. Implicit feedback is noisy. While tracking user behavior, we can only guess the users preferences and motives. For example a user may do a ride in Golden, CO often but this does not necessarily indicate a positive view of the activity. They may only do it because other more prefered rides require a drive or have more danger presented by traffic.
    
3. Numerical signal in explicit feedback represents preference whereas, with implicit feedback it represents confidence. With implicit feedback, a numerical value indicates frequency of an action. A higher value may indicate higher confidence in the users decision to "consume" or perform an activity but doesn't necessarily mean they would have higher preference for that item. They may do a ride once and absolutely love it. Or, they may ride lookout mountain many times, but mostly out of desire to become a better road climber.
    
4. Evaluation for implicit feedback recommenders requires different evaluation metrics. We have to consider availability of an activity, competition for the activity with other activities and repeat feedback.

Baseline methods provide some context for comparing our recommender algorithm against those of little computational complexity. It's difficult to claim we have good results without some known reference point. We’ll define three very simple baseline methods and find the RMSE using these methods. Our goal will be to obtain lower RMSE scores with whatever model we produce.

The first method will be **Uniform Random Baseline**. Everywhere we see a blank rating, we will fill it in with a number drawn uniformly from the range 0 to max(ratings).

The second will be **Global Mean Baseline**. This method is only slightly better than the last. Wherever we have a missing value, we’ll fill it in with the mean of all observed ratings.

Finally, the third is **Mean of Means Baseline**. We imagine some users might be more apt to do a ride that's close. So they might "rate" these activities better than others. Others may be they opposite, prefering to do a different activity every time. Additionally, some rides may simply be better than others. We’ll attempt to capture these general trends through per-user and per-joke rating means. We’ll also incorporate the global mean to smooth things out a bit. So if we see a missing value in cell Rij, we’ll average the global mean with the mean of Ui and the mean of Vj and use that value to fill it in.

Baseline methods are stored in an external file called baselines.py

```python
from baselines import *
import hashlib
import os
DATA_DIR = '../data/baseline'
```

```python
baseline_methods = OrderedDict()
baseline_methods['ur'] = UniformRandomBaseline
baseline_methods['gm'] = GlobalMeanBaseline
baseline_methods['mom'] = MeanOfMeansBaseline
```

Create the utility matrix and fill in missing values using above described methods.

```python
activity_ids = np.sort(df.label.unique())
athlete_ids = np.sort(df.athlete_id.unique())
utility = np.empty((len(athlete_ids), len(activity_ids)))
utility[:] = np.nan

for k, v in ratings_dict.iteritems():
    utility[np.searchsorted(athlete_ids, k), np.searchsorted(activity_ids, v[0])] = v[1]
```

Check the sparsity of our data. This is equivalent to determining the number of ratings vs. the total number of possible values in the matrix.

```python
sparsity = float(np.count_nonzero(~np.isnan(utility)))
sparsity /= (utility.shape[0] * utility.shape[1])
sparsity *= 100
print 'Sparsity: {:4.2f}%'.format(sparsity)
```

```python
test_np = utility

train_np = utility.copy()

rating_indices = np.argwhere(utility > 0)
size = rating_indices.shape[0]
rows = np.random.choice(size, int(size * .1), replace=False)
mask_idxs = rating_indices[rows]

for row in mask_idxs:
    train_np[tuple(row)] = np.nan
```

```python
# Let's see the results:
baselines = {}
for name in baseline_methods:
    Method = baseline_methods[name]
    method = Method(train_np)
    baselines[name] = method.rmse(test_np)
    print('%s RMSE:\t%.5f' % (method, baselines[name]))
```

The first method, Uniform Random Baseline, yields some predictably bad results. The Global Mean Baseline method also does not achieve a very good RMSE. The Mean of Means baseline method does not even work and simply outputs a nan value. One possible reason for this is the sparsity of the dataset.

```python
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.recommendation import ALS
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
from pyspark.ml import Pipeline
from pyspark.sql import Row
from pyspark.sql.functions import col

lines = spark.read.text("../data/ratings.txt").rdd
parts = lines.map(lambda row: row.value.split("::"))
ratingsRDD = parts.map(lambda p: Row(userId=int(p[0]), item=int(p[1]),
                                     rating=float(p[2])))
ratings = spark.createDataFrame(ratingsRDD)
(training_df, test_df) = ratings.randomSplit([0.8, 0.2])

# Build the recommendation model using ALS on the training data
als = ALS(maxIter=5, regParam=0.01, userCol="userId", itemCol="item", ratingCol="rating", implicitPrefs=True)
model = als.fit(training_df)

# Evaluate the model by computing the RMSE on the test data
predictions = model.transform(test_df)
# evaluator = RegressionEvaluator(metricName="rmse", labelCol="rating",
#                                 predictionCol="prediction", coldStartStrategy="drop")
# rmse = evaluator.evaluate(predictions)
# print("Root-mean-square error = " + str(rmse))
```
