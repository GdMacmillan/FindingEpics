Read in data from the toy dataset, taking only activities of type run, in the state of Colorado, that are not commutes. Also, drop all rows missing a summary polyline and reset the index of the dataframe. This is required for the clustering operations to work.

```python
df = pd.read_csv('../data/activities_small.csv', encoding='utf-8')
df = df[df['commute'] == 0] # drop all where commute = true
df = df[df['state'] == 'Colorado'] # drop all where state != Colorado
df = df[df['type'] == 'Ride'] # drop everything that is not of type 'Ride'
df = df.dropna(subset=['map_summary_polyline']).reset_index(drop=True) # drop all where
```
